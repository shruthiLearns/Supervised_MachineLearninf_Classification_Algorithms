# <span style="color:red">LOGISTIC REGRESSION: A  TYPE OF BINARY/ BINOMIAL CLASSIFICATION</SPAN>

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
```

#### <span style="color:BLUE"> ACCESSING THE DATA</SPAN>

```python
url = "https://raw.githubusercontent.com/DataScience75/Batch-114/refs/heads/main/Students.csv"
data = pd.read_csv(url)
print("Number of records : ",data.shape[0])
print("Number of Columns : ",data.shape[1])
print("\n",data.head())
```

### <span style="color:BLUE">PLOT THE DATA USING THE SCATTER PLOT</SPAM>

```python
sns.scatterplot(data = data, x = "JEEPRE", y="JEEMAINS",hue='STATUS') #,palette='winter'
plt.show()
```

#### <span style="color:BLUE">SELECTING THE IMPORTANT FEATURES

```python
X=data.iloc[:,0:2]
y=data.iloc[:,2]
```

```python
X.shape,y.shape
```

#### <span style="color:BLUE">SAMPLING AND SPLITTING THE DATA INTO TRAINING AND TESTTING DATASET

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3)
```

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

### <span style="color:BLUE">CREATE A LOGISTIC REGRESSION - A BINARY CLASSIFICATION MODEL 

```python
from sklearn.linear_model import LogisticRegression
```

#### Create an Object of type LogisticRegression

```python
log_reg_model = LogisticRegression()
```

```python
print(type(log_reg_model))
```

### <span style="color:BLUE">Training the Logistic Regression Model: (fit) the Logistic regression Model

```python
%%time
log_reg_model_train = log_reg_model.fit(X_train,y_train)
print("Logistic Regression Model Training Done")
```

### <span style="color:BLUE"> CALCULATE THE PREDICTED PROBABILITY FOR BOTH THE CLASSES (0 & 1)

```python
predicted_probability  = log_reg_model_train.predict_proba(X_test)
print(type(predicted_probability))
print("Dimension of the array = ",predicted_probability.ndim)
print("Size of the array", len(predicted_probability)) # Should be the same as that of the X_test recor count 30
predicted_probability
```

### <span style="color:BLUE"> CALCULATE THE PREDICTED OUTPUT

```python
y_pred = log_reg_model_train.predict(X_test)
```

```python
y_pred
```

### <span style="color:BLUE"> CREATE A NEW DATAFRAME TO SEE THE ACTUAL OUTPUT AND PREDICTED OUTPUT

```python
y_test_actual = y_test.values.reshape(1,-1).flatten()
y_test_actual
```

```python
y_test_actual.ndim,y_pred.ndim
```

```python
actualVsPredicted = pd.DataFrame({"Actual_output":y_test_actual,"Predicted_output":y_pred})
```

```python
actualVsPredicted
```

### <span style="color:BLUE"> CALCULATE AND PLOT CONFUSION MARTIX <br><br>TO EVALUATE THE PERFORMACE OF THE LOGISTIC REGRESSION CLASSIFICATION MODEL

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
```

```python
confusionMatrix = confusion_matrix(y_test,y_pred)
confusionMatrix
```

```python
# True Positive = 9
# True Nagetive = 18
# False Positive = 3
# False Nagetive = 0
```

#### <span style="color:BLUE">DISPLAYING THE CONFUSION MATRIX

```python
conf_matrix_disp = ConfusionMatrixDisplay(confusionMatrix, display_labels = log_reg_model.classes_)
conf_matrix_disp.plot()
```

**<span style="color:BLUE">(.classes_) which is an attribute commonly found in machine learning models from libraries like scikit-learn. <br>This attribute stores the unique class labels that a classifier has learned during training.**

```python
print(log_reg_model.classes_)
```

#### <span style="color:BLUE">CALCULATE THE ACCURACY OF THE MODEL

```python
accuracy = (9+18)/(9+3+0+18)
print(f"The Accuracy of the Logistic Model is {accuracy*100}%")
```

#### <span style="color:BLUE">CALCULATE THE ERROR OF THE MODEL (1-accuarcy)

```python
error_rate = 1-accuracy
print(f"The Error of the Logistic Model is {error_rate*100}%")
```
